```python
import pandas as pd
df = pd.read_csv("model/common/data/categories_input.csv")
df
```

```python
import json
labels = json.loads(open("./model/common/data/labels.json", "r").read())
labels
```

```python
assert len(set(labels)) == len(labels)
assert len(set(labels) - set(df["ICD Block Name"]))==0
print("Those block names were discarded:\n" + "\n".join(set(df["ICD Block Name"]) - set(labels)))
```

```python
df = df[df["ICD Block Name"].isin(labels)]
df
```

```python
df = df.rename(columns = {
    'ICD Block Name':'block_name',
    'ICD Chapter Name': 'chapter_name',
    'ICD Block Code': 'block_code',
    'ICD Chapter Code': 'chapter_code'
})
df
```

```python
df["link"] = [f"https://icd.who.int/browse10/2019/en#/{code}" for code in df["block_code"]]
```

```python
df.to_csv("model/common/data/categories.csv", index=False)
```
